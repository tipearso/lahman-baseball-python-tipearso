```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/lahman_baseball")
```

```python
people = pd.read_sql("SELECT * FROM people;", con=engine)

c_people =pd.read_sql("SELECT * FROM collegeplaying", con=engine)
c_people.head()
schools =pd.read_sql("SELECT * FROM schools", con=engine)
salaries = pd.read_sql("SELECT * FROM salaries", con=engine)
```

```python
schools =pd.read_sql("SELECT * FROM schools", con=engine)
schools.head()
```

```python
vandy = schools.loc[schools['schoolname']== 'Vanderbilt University']
# college_people.sort_values()
vandy
```

```python
vandy.info()
```

combined DF to get college players who went to vanderbilt

```python
v_cp = pd.merge(c_people,vandy)
v_cp2 =  v_cp[['schoolid','playerid']]
v_cp_clean = v_cp2.drop_duplicates()
v_cp_clean
```

combine this with people DF

```python
vandy_players = pd.merge(people, v_cp_clean, on ='playerid')
vandy_players
vandy_salary = pd.merge(vandy_players, salaries, on ='playerid')
vandy_salary
```

```python
salary_name = vandy_salary[['namefirst','namelast','playerid','salary']]
salary_name
```

```python
#I feel fairly certain there is a much faster way to do all of what I just did... but I did it. 
```

```python
salary_grouped = salary_name.groupby(['namefirst','namelast','playerid'],as_index = False).sum()
salary_vandy = salary_grouped.sort_values('salary', ascending =False)
salary_vandy = salary_vandy.drop(['playerid'], axis = 1).reset_index(drop = True)
                                
salary_vandy
```
